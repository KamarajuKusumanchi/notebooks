```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(5,3), columns=list('abc')) - 0.5
df
```

```python
df.iloc[df['a'].abs().argsort()]
```

```python
# To reset the index
df.iloc[df['a'].abs().argsort()].reset_index(drop=True)
```

```python
# To sort in descending order negate df['a'].abs()
df.iloc[-df['a'].abs().argsort()]
```
