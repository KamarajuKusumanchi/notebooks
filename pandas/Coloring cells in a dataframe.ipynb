```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three',
                          'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8),
                   'E' : np.random.randn(8)})
print(df)
```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

# The above function does not work on strings. 
# So subsetting on columns with numbers instead.
df2 = df[['C', 'D']]
s = df2.style.applymap(color_negative_red)
s
```

```python
def color_string_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for some
    predefined strings, black otherwise.
    """
    color = 'red' if val in ('foo', 'two') else 'black'
    return 'color: %s' % color
s = df.style.applymap(color_string_red)
s
```
